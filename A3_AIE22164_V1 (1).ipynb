```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "C:\\Users\\91914\\Downloads\\Lab_Session1_Data(1).xlsx" # Providing the path for the excel sheet 1
DataFrame = pd.read_excel(file_path).iloc[:, :5] # Read the Excel file into a DataFrame
print(DataFrame)

A_matrix = DataFrame.iloc[:, 1:4]  # Matrix A contains columns 1 through 3
C_vector = DataFrame.iloc[:, 4]    # Vector C contains the last column
print("\n A matrix: ")
print(A_matrix) # Print Matrix A
print("\n C vector: ") # Print Matrix C
print(C_vector)

vector_dimension = A_matrix.shape[1] # Calculating and print the dimensionality of the vector space
print("Dimensionality of the vector space:", vector_dimension)
vector_space = A_matrix.shape[0] # Calculating and print the number of vectors in the vector space
print("Number of vectors in the vector space:", vector_space)

rank_A = np.linalg.matrix_rank(A_matrix) # Calculating and print the rank of matrix A
print("Rank of Matrix A:", rank_A)

pseudo_inv_A = np.linalg.pinv(A_matrix) # Calculating the pseudo-inverse of matrix A
AinverseC = np.dot(pseudo_inv_A, C_vector) # Calculating the product of each
print(f"Cost of each product: {AinverseC}") # Print the Vector

X = np.dot(pseudo_inv_A, C_vector) # Calculating the model vector X for predicting the cost of the products
print("Vector X for predicting the cost of the products: ") # Print the model vector X
print(X)



DataFrame['Class'] = np.where(DataFrame['Payment (Rs)'] > 200, 'RICH', 'POOR') # Mark customers as RICH or POOR based on payments
print("\nUpdated DataFrame:")
print(DataFrame) # Print the DataFrame with the changes


print("\n \n  -----------------------------------------IRTC ----------------------------------------------")
DataFrame1 = pd.read_excel('C:\\Users\\91914\\Downloads\\Lab_Session1_Data(1).xlsx', sheet_name=1) # Providing the path for the excel sheet 2
print(f"\n {DataFrame1}") # Print the DataFrame


mean_D = DataFrame1['Price'].mean() # Calculating the mean and variance of the 'Price' column
variance_D = DataFrame1['Price'].var()
print('Mean:', mean_D) # Print the Mean
print('Variance:', variance_D) # Print the Variance


wednesday_df = DataFrame1[DataFrame1['Day'] == 'Wed'] # Selecting the rows where the day of the week is Wednesday
wednesday_mean = wednesday_df['Price'].mean() # Calculating the mean of the 'Price' column for these rows
population_mean = DataFrame1['Price'].mean() # Calculate the mean of the 'Price' column for all rows (population mean)
print('Wednesday Mean:', wednesday_mean) # Print Wednesdays mean
print('Population Mean:', population_mean) # Print Population mean


April_df = DataFrame1[DataFrame1['Month'] == 'Apr'] # Selecting the rows where the month is April
April_mean = April_df['Price'].mean() # Calculating the mean of the 'Price' column for April


population_mean = DataFrame1['Price'].mean() # Calculating the population mean
print('April Mean:', April_mean)  # Print April mean
print('Population Mean:', population_mean)  # Print Population mean


l2 = list(map(lambda v: v < 0, DataFrame1['Chg%'])) # Calculating probability of 'Chg%' values being negative
l2_false = [value for value in l2 if value is False]
probability = (len(l2_false) / len(l2))
l0 = 1 - probability
print(f'Probability: {probability}%')  # Print Probability


l3 = list(map(lambda v: v > 0, wednesday_df['Chg%'])) # Calculating probability of 'Chg%' values being positive on Wednesday
l3_True = [value for value in l3 if value is True]
probability_wed = (len(l3_True) / len(l3))
conditional_prob = probability_wed * l0
print(f'Profits on Wednesday: {probability_wed}%')
print(f'Conditional Probability: {conditional_prob}%')


sns.scatterplot(x='Day', y='Chg%', data=DataFrame1) # Create a scatter plot of 'Chg%' data against the day of the week
plt.show() # Display the plot
```
